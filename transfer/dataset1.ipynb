```python
import os
mu = os.listdir('/data/GZ_Decals/merge')
print(len(mu))
dd = []
for i in mu:
    tep = i.split('_')[:2]
    try:
        tep = [eval(j) for j in tep]
        tep.append('/data/GZ_Decals/merge/'+i)
    except:
        continue
    dd.append(tep)
mu = os.listdir('/data/GZ_Decals/nomerge')
for i in mu:
    tep = i.split('_')[:2]
    try:
        tep = [eval(j) for j in tep]
        tep.append('/data/GZ_Decals/nomerge/'+i)
    except:
        continue
    dd.append(tep)
print(len(mu))
```

```python
print(len(dd))
data1 = pd.DataFrame(dd)
data1.head()
```

```python
from tqdm import tqdm
import numpy as np
dd = []
for j in tqdm(range(len(ain))):
    da = data1[(data1[data1.columns[0]]==ain.iloc[j,1]) & (data1[data1.columns[1]]==ain.iloc[j,2])]
    if len(da) ==  1:
        da1 = da
    else:
        da = data1[(abs(data1[data1.columns[0]]-ain.iloc[j,1])<1e-4) & (abs(data1[data1.columns[1]]-ain.iloc[j,2])<1e-4)]
        if len(da) !=  0:
            da1 = da
        # da = data1[(abs(data1[data1.columns[0]]-ain.iloc[j,1])<1e-4) & (abs(data1[data1.columns[1]]-ain.iloc[j,2])<1e-6)]
        else:
            continue
    re = da1.index.item()
    if type(re)!=int:
        re = re[0]
    re1 = np.hstack((data1.iloc[re,2],re,j))
    dd.append(re1)
dd = np.array(dd)
```

```python
adre = pd.DataFrame(dd)
adre.columns = ['address','rank1','rank2']
adre.head()
```
