```python
import os
mu = os.listdir('/data/GZ_Decals/MGS_out_DECaLS')
print(len(mu))
dd = []
for i in mu:
    tep = i.split('_')[:2]
    try:
        tep = [eval(j) for j in tep]
        tep.append(i)
    except:
        continue
    dd.append(tep)
print(dd[:10])    
```

```python
data1 = pd.DataFrame(dd)
data1.head()
```

```python
from tqdm import tqdm
import numpy as np
dd = []
for j in tqdm(range(len(ain))):
    da = data1[(data1[data1.columns[0]]==ain.iloc[j,1]) & (data1[data1.columns[1]]==ain.iloc[j,2])]
    if len(da) ==  1:
        da1 = da
    else:
        da = data1[(abs(data1[data1.columns[0]]-ain.iloc[j,1])<1e-4) & (abs(data1[data1.columns[1]]-ain.iloc[j,2])<1e-4)]
        if len(da) !=  0:
            da1 = da
        # da = data1[(abs(data1[data1.columns[0]]-ain.iloc[j,1])<1e-4) & (abs(data1[data1.columns[1]]-ain.iloc[j,2])<1e-6)]
        else:
            continue
    re = da1.index.item()
    re1 = np.hstack((data1.iloc[re,2],re,j))
    dd.append(re1)
dd = np.array(dd)
```

```python
print(dd[:10])
print(len(dd))
# print(j)
# print(len(da1))
# print(da)
# print(da1)
```

```python
adre = pd.DataFrame(dd)
adre.columns = ['address','rank1','rank2']
adre.head()
```

```python
dat = pd.concat([ain,adre],axis=1)
dat.head()
```

```python
dat.to_csv('result.csv')
```
