```python
import pandas as pd
ain = pd.read_csv('resu.csv')
print(ain.shape)
ain.head()
```

```python
import os
from astropy.io import fits

dir = '/data/GZ_Decals/MGS_out_DECaLS/'
train = open('train_t.txt', 'w')

for b in ain['address_1']:
    ss = dir + b
    try:
        name = ss + '\n'
        hdu = fits.open(ss)
        img = hdu[0].data
        if len(img)==3 and len(img[0])==256:
            hdu.close()
            train.write(name)
    except:
        continue

train.close()
```
